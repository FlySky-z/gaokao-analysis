```python
import pandas as pd
import json

province_list = [
    '北京', '天津', '上海', '重庆', '河北', '山西', '辽宁', '吉林', '黑龙江',
    '江苏', '浙江', '安徽', '福建', '江西', '山东', '河南', '湖北', '湖南',
    '广东', '海南', '四川', '贵州', '云南', '陕西', '甘肃', '青海', '台湾',
    '内蒙古', '广西', '西藏', '宁夏', '新疆', '香港', '澳门'
]

def split_province_city(row):
    city_str = str(row).strip()
    for prov in province_list:
        if city_str.startswith(prov):
            city = city_str[len(prov):]
            return prov, city
    return None, city_str  # 省份未知

def province_city_dict(df, col_name):
    result = {}
    for val in df[col_name].dropna():
        prov, city = split_province_city(val)
        if not prov:
            continue
        if prov not in result:
            result[prov] = []
        # 去重
        if city not in result[prov]:
            result[prov].append(city)
    return result

province_city_json = province_city_dict(df, "院校基础信息_城市")
json_str = json.dumps(province_city_json, ensure_ascii=False, indent=2)
print(json_str)
with open("province_city.json", "w", encoding="utf-8") as f:
    json.dump(province_city_json, f, ensure_ascii=False, indent=2)
```

```python
# 处理 排名数据
# 读取一分一段表数据
import json

# 读取历史类和物理类一分一段表数据
with open("../web/data/ranking_score_hubei_history.json", 'r', encoding='utf-8') as f:
    history_ranking = json.load(f)

with open("../web/data/ranking_score_hubei_physics.json", 'r', encoding='utf-8') as f:
    physics_ranking = json.load(f)

print("成功读取一分一段表数据")
print(f"历史类数据条数: {len(history_ranking['data'])}")
print(f"物理类数据条数: {len(physics_ranking['data'])}")

# 查看数据结构
print("\n历史类前5条数据:")
for i in range(5):
    print(history_ranking['data'][i])

print("\n物理类前5条数据:")
for i in range(5):
    print(physics_ranking['data'][i])
```

```python
# 创建分数到位次的映射函数
def create_score_rank_mapping(ranking_data):
    """
    创建分数到位次的映射字典
    """
    score_rank_map = {}
    
    for item in ranking_data['data']:
        score_str = item['score']
        accumulate = item['accumulate']
        
        # 处理分数范围（如"695-750"）和单一分数（如"694"）
        if '-' in score_str:
            # 分数范围，取最低分
            min_score = int(score_str.split('-')[0])
            # 对于范围分数，我们将最低分映射到累计位次
            score_rank_map[min_score] = accumulate
        else:
            # 单一分数
            try:
                score = int(score_str)
                score_rank_map[score] = accumulate
            except ValueError:
                # 跳过无法转换为整数的分数
                continue
    
    return score_rank_map

# 创建历史类和物理类的分数位次映射
history_score_rank = create_score_rank_mapping(history_ranking)
physics_score_rank = create_score_rank_mapping(physics_ranking)

print("历史类分数位次映射创建完成")
print(f"历史类映射条数: {len(history_score_rank)}")
print("历史类部分映射示例:")
for i, (score, rank) in enumerate(list(history_score_rank.items())[:5]):
    print(f"分数 {score} -> 位次 {rank}")

print("\n物理类分数位次映射创建完成")
print(f"物理类映射条数: {len(physics_score_rank)}")
print("物理类部分映射示例:")
for i, (score, rank) in enumerate(list(physics_score_rank.items())[:5]):
    print(f"分数 {score} -> 位次 {rank}")
```

```python
# 创建根据分数查找位次的函数
def get_rank_by_score(score, score_rank_map):
    """
    根据分数查找对应的位次
    如果找不到精确分数，返回最接近的较低分数对应的位次
    """
    if pd.isna(score) or score is None:
        return None
    
    try:
        score = int(float(score))  # 确保分数是整数
    except (ValueError, TypeError):
        return None
    
    # 首先尝试精确匹配
    if score in score_rank_map:
        return score_rank_map[score]
    
    # 如果找不到精确匹配，找到最接近的较低分数
    available_scores = sorted(score_rank_map.keys(), reverse=True)
    for available_score in available_scores:
        if available_score <= score:
            return score_rank_map[available_score]
    
    # 如果分数太低，返回最低分数对应的位次
    if available_scores:
        return score_rank_map[min(available_scores)]
    
    return None

def get_rank_by_score_and_subject(score, subject_type):
    """
    根据分数和科类获取对应位次
    """
    if subject_type == '物理':
        return get_rank_by_score(score, physics_score_rank)
    elif subject_type == '历史':
        return get_rank_by_score(score, history_score_rank)
    else:
        # 如果科类不明确，默认使用物理类映射
        return get_rank_by_score(score, physics_score_rank)

# 测试函数
print("测试根据科类匹配位次的函数:")
test_data = [
    (650, '物理'),
    (600, '历史'),
    (550, '物理'),
    (500, '历史')
]

for score, subject in test_data:
    rank = get_rank_by_score_and_subject(score, subject)
    print(f"分数 {score}, 科类 {subject} -> 位次 {rank}")
```

```python
# 根据专业最低分和科类填写专业最低位次
def fill_ranks_by_scores_and_subject(df):
    """
    根据专业最低分和科类填写专业最低位次
    """
    # 创建副本避免修改原DataFrame
    df_copy = df.copy()
    
    # 检查必要的列是否存在
    score_col = '2024年录取数据_专业最低分'
    rank_col = '2024年录取数据_专业最低位次'
    subject_col = '2024年计划数据_科类'
    
    required_cols = [score_col, rank_col, subject_col]
    missing_cols = [col for col in required_cols if col not in df_copy.columns]
    
    if missing_cols:
        print(f"错误: 缺少必要的列 {missing_cols}")
        return df_copy
    
    print(f"开始处理 {len(df_copy)} 条记录...")
    print(f"使用科类列: {subject_col}")
    
    # 统计变量
    updated_count = 0
    history_count = 0
    physics_count = 0
    no_score_count = 0
    no_subject_count = 0
    
    # 遍历每行数据
    for index, row in df_copy.iterrows():
        score = row[score_col]
        subject_type = row[subject_col]
        
        # 跳过没有分数的记录
        if pd.isna(score) or score is None:
            no_score_count += 1
            continue
            
        # 跳过没有科类信息的记录
        if pd.isna(subject_type) or subject_type is None or subject_type == '':
            no_subject_count += 1
            continue
        
        # 直接使用整数分数，不进行float转换
        try:
            score = int(score)
        except (ValueError, TypeError):
            no_score_count += 1
            continue
        
        # 根据科类获取对应的位次，直接覆盖原有值
        final_rank = get_rank_by_score_and_subject(score, subject_type)
        
        if final_rank is not None:
            df_copy.at[index, rank_col] = final_rank
            updated_count += 1
            
            # 统计各科类的更新数量
            if subject_type == '历史':
                history_count += 1
            elif subject_type == '物理':
                physics_count += 1
        
        # 每1000行显示一次进度
        if (index + 1) % 1000 == 0:
            print(f"已处理 {index + 1} 行...")
    
    print(f"\n处理完成!")
    print(f"总记录数: {len(df_copy)}")
    print(f"成功更新位次: {updated_count} 条")
    print(f"历史类更新: {history_count} 条")
    print(f"物理类更新: {physics_count} 条")
    print(f"无分数记录: {no_score_count} 条")
    print(f"无科类记录: {no_subject_count} 条")
    
    return df_copy

# 执行位次填写
print("开始根据分数和科类填写位次...")
df_updated = fill_ranks_by_scores_and_subject(df)

# 显示更新结果
print("\n更新前后对比 (前10行):")
comparison_cols = ['2024年计划数据_科类', '2024年录取数据_专业最低分', '2024年录取数据_专业最低位次']
if all(col in df.columns for col in comparison_cols):
    print("原始数据前10行:")
    print(df[comparison_cols].head(10))
    print("\n更新后数据前10行:")
    print(df_updated[comparison_cols].head(10))
```
