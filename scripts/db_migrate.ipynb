```python
import pandas as pd

def find_key_columns(df, keywords):
    """
    查找包含特定关键词的列名
    :param df: DataFrame
    :param keywords: 关键词列表
    :return: 包含关键词的列名列表
    """
    key_columns = []
    for col in df.columns:
        if any(keyword in col for keyword in keywords):
            key_columns.append(col)

    print("院校相关列:")
    for i, col in enumerate(key_columns):
        print(f"{i+1}: {col}")

    # 取每个关键列的前2个值，构建纵向表格
    preview_data = []
    for col in key_columns:
        values = df[col].head(2).tolist()
        preview_data.append([col] + values)

    # 构建DataFrame，第一列为列名，后面为前2个值
    preview_vertical_df = pd.DataFrame(preview_data, columns=['列名', '样例1', '样例2'])
    display(preview_vertical_df)

college_key_words = ['招生计划']
find_key_columns(df, college_key_words)
```

```python
# 选科要求字段解析
def check_subject(subject, row):
    # 检查选科要求原始字段是否包含该学科
    if subject in str(row['subject_requirement_raw']):
        return True
    # 检查科类字段是否包含该学科
    if subject in str(row['subject_category']):
        return True
    return False

cleaned_df['require_physics'] = cleaned_df.apply(lambda row: check_subject('物', row), axis=1)
cleaned_df['require_chemistry'] = cleaned_df.apply(lambda row: check_subject('化', row), axis=1)
cleaned_df['require_biology'] = cleaned_df.apply(lambda row: check_subject('生', row), axis=1)
cleaned_df['require_politics'] = cleaned_df.apply(lambda row: check_subject('政', row), axis=1)
cleaned_df['require_history'] = cleaned_df.apply(lambda row: check_subject('历', row), axis=1)
cleaned_df['require_geography'] = cleaned_df.apply(lambda row: check_subject('地', row), axis=1)

print("选科要求字段解析完成")
print("选科要求统计:")
print(f"要求物理: {cleaned_df['require_physics'].sum()}")
print(f"要求化学: {cleaned_df['require_chemistry'].sum()}")
print(f"要求生物: {cleaned_df['require_biology'].sum()}")
print(f"要求政治: {cleaned_df['require_politics'].sum()}")
print(f"要求历史: {cleaned_df['require_history'].sum()}")
print(f"要求地理: {cleaned_df['require_geography'].sum()}")
print("【专业】字段解析全部完成")
cleaned_df.head()
```

```python
import json

def load_score_rank(year, category):
    """
    根据规则读取数据
    """
    # category: 'physics' or 'history'
    category_map = {'physics': '物理', 'history': '历史'}
    filename = f'../static/score_rank_hubei_{year}_{category}.json'
    with open(filename, 'r', encoding='utf-8') as f:
        score_rank = json.load(f)
    subject = category_map[category]
    score_rank_map = {int(item['score']): int(item['accumulate']) for item in score_rank["data"]}
    max_score = max(score_rank_map.keys())
    score_rank_map = {k: v for k, v in score_rank_map.items()}
    # 额外记录最高分
    score_rank_map['max_score'] = str(max_score)
    return f"{year}_{subject}", score_rank_map

# 科目-分数-位次映射
subject_score_rank_map = {}
for cat in ['physics', 'history']:
    key, score_rank_map = load_score_rank(2024, cat)
    subject_score_rank_map[key] = score_rank_map

# 简要查看subject_score_rank_map内容
for subject_key, score_rank in subject_score_rank_map.items():
    print(f"{subject_key}: 共{len(score_rank)-1}个分数-位次映射，分数范围: {min(int(k) for k in score_rank if k != 'max_score')}~{score_rank['max_score']}")
    preview = list(score_rank.items())[:5]
    print("  示例:", preview, " 最高分",score_rank['max_score'])
```

```python
# 统一tag字段的分隔方式
def unify_tag_columns(tag):
    # 将标签字段的分隔符统一为英文逗号
    if isinstance(tag, str):
        tag = tag.strip()
        tag = tag.replace('，', ',')
        tag = tag.replace('/', ',')
        tag = tag.replace(' ', ',')
        # 去除重复逗号和首尾逗号
        tag = ','.join([t for t in tag.split(',') if t.strip()])
    return tag
```

执行列数据分析

```python
def analyze_and_generate_clickhouse_schema(df, enum_threshold=16, string_length_threshold=255):
    """
    自动分析DataFrame并生成ClickHouse建表语句
    
    Parameters:
    df: pandas.DataFrame - 要分析的数据框
    enum_threshold: int - 唯一值数量阈值，低于此值的字符串列会使用Enum类型
    string_length_threshold: int - 字符串长度阈值，超过此长度使用Text类型
    """
    
    schema_info = []
    
    for col_name in df.columns:
        col_data = df[col_name]
        dtype = col_data.dtype
        
        # 统计基本信息
        null_count = col_data.isnull().sum()
        total_count = len(col_data)
        null_ratio = null_count / total_count
        unique_count = col_data.nunique()
        
        # 判断是否需要Nullable
        nullable = null_ratio > 0
        
        print(f"\n分析列: {col_name}")
        print(f"  数据类型: {dtype}")
        print(f"  空值数量: {null_count} ({null_ratio:.2%})")
        print(f"  唯一值数量: {unique_count}")
        
        # 根据数据类型判断ClickHouse类型
        if pd.api.types.is_string_dtype(dtype) or dtype == 'object':
            # 字符串类型处理
            non_null_data = col_data.dropna()
            
            if len(non_null_data) == 0:
                ch_type = 'String'
            elif unique_count <= enum_threshold:
                # 使用Enum类型
                unique_values = non_null_data.unique()
                # 过滤掉过长的值，避免Enum类型过于复杂
                valid_values = [str(v) for v in unique_values if len(str(v)) <= 100]
                
                if len(valid_values) == unique_count and len(valid_values) <= enum_threshold:
                    # 为每个值分配一个数字ID
                    enum_items = [f"'{v}' = {i+1}" for i, v in enumerate(valid_values)]
                    ch_type = f"Enum8({', '.join(enum_items)})"
                    print(f"  -> 使用Enum8类型，包含{len(valid_values)}个值")
                else:
                    ch_type = 'String'
            else:
                # 检查字符串长度
                max_length = non_null_data.astype(str).str.len().max() if len(non_null_data) > 0 else 0
                if max_length > string_length_threshold:
                    ch_type = 'Text'
                    print(f"  -> 使用Text类型，最大长度: {max_length}")
                else:
                    ch_type = 'String'
                    print(f"  -> 使用String类型，最大长度: {max_length}")
        
        elif pd.api.types.is_integer_dtype(dtype):
            # 整数类型处理
            non_null_data = col_data.dropna()
            
            if len(non_null_data) == 0:
                ch_type = 'Int32'
            else:
                min_val = non_null_data.min()
                max_val = non_null_data.max()
                
                print(f"  -> 数值范围: {min_val} ~ {max_val}")
                
                # 判断是否有负数
                has_negative = min_val < 0
                
                # 根据数值范围选择合适的整数类型
                if has_negative:
                    if min_val >= -128 and max_val <= 127:
                        ch_type = 'Int8'
                    elif min_val >= -32768 and max_val <= 32767:
                        ch_type = 'Int16'
                    elif min_val >= -2147483648 and max_val <= 2147483647:
                        ch_type = 'Int32'
                    else:
                        ch_type = 'Int64'
                else:
                    if max_val <= 255:
                        ch_type = 'UInt8'
                    elif max_val <= 65535:
                        ch_type = 'UInt16'
                    elif max_val <= 4294967295:
                        ch_type = 'UInt32'
                    else:
                        ch_type = 'UInt64'
        
        elif pd.api.types.is_float_dtype(dtype):
            # 浮点数类型处理
            ch_type = 'Float64'
            
        elif pd.api.types.is_bool_dtype(dtype):
            # 布尔类型处理
            ch_type = 'Bool'
            
        elif pd.api.types.is_datetime64_any_dtype(dtype):
            # 日期时间类型处理
            ch_type = 'DateTime'
            
        else:
            # 默认类型
            ch_type = 'String'
            print(f"  -> 未识别的数据类型，使用默认String类型")
        
        # 添加Nullable包装
        if nullable and ch_type not in ['String', 'Text']:  # String类型默认可以为空
            ch_type = f'Nullable({ch_type})'
        
        print(f"  -> ClickHouse类型: {ch_type}")
        
        # 保存schema信息
        schema_info.append({
            'column_name': col_name,
            'pandas_dtype': str(dtype),
            'clickhouse_type': ch_type,
            'null_count': null_count,
            'null_ratio': null_ratio,
            'unique_count': unique_count
        })
        
    return schema_info

def override_column_types(schema_info, type_overrides):
    """
    覆盖指定列的数据类型
    
    Parameters:
    schema_info: list - 从analyze_and_generate_clickhouse_schema返回的schema信息
    type_overrides: dict - 列名到新类型的映射, 例如 {'column_name': 'NewType'}
    
    Returns:
    list - 更新后的schema信息
    """
    updated_schema = []
    
    for col_info in schema_info:
        col_name = col_info['column_name']
        
        # 创建新的列信息副本
        new_col_info = col_info.copy()
        
        # 检查是否需要覆盖类型
        if col_name in type_overrides:
            old_type = col_info['clickhouse_type']
            new_type = type_overrides[col_name]
            
            print(f"覆盖列 '{col_name}': {old_type} -> {new_type}")
            new_col_info['clickhouse_type'] = new_type
            new_col_info['type_overridden'] = True
        else:
            new_col_info['type_overridden'] = False
        
        updated_schema.append(new_col_info)
    
    return updated_schema

def print_schema_summary(schema_info):
    """
    打印schema摘要信息
    """
    print("Schema摘要:")
    print("-" * 80)
    
    total_columns = len(schema_info)
    overridden_columns = sum(1 for col in schema_info if col.get('type_overridden', False))
    
    print(f"总列数: {total_columns}")
    print(f"类型被覆盖的列数: {overridden_columns}")
    
    # 按类型分组统计
    type_count = {}
    for col in schema_info:
        col_type = col['clickhouse_type']
        # 简化类型名（去掉参数）
        simple_type = col_type.split('(')[0]
        type_count[simple_type] = type_count.get(simple_type, 0) + 1
    
    print("\n类型分布:")
    for col_type, count in sorted(type_count.items()):
        print(f"  {col_type}: {count}")
    
    # 显示被覆盖的列
    if overridden_columns > 0:
        print("\n被覆盖类型的列:")
        for col in schema_info:
            if col.get('type_overridden', False):
                print(f"  {col['column_name']}: {col['clickhouse_type']}")


def generate_create_table_from_schema(schema_info, table_name, order_by_columns=None, engine='MergeTree'):
    """
    根据schema信息生成ClickHouse建表语句
    
    Parameters:
    schema_info: list - 列schema信息
    table_name: str - 表名
    order_by_columns: list - 排序列，如果为None则使用tuple()
    engine: str - 表引擎，默认为MergeTree
    
    Returns:
    str - 完整的CREATE TABLE语句
    """
    
    # 生成列定义
    column_definitions = []
    for col_info in schema_info:
        col_name = col_info['column_name']
        col_type = col_info['clickhouse_type']
        column_definitions.append(f"    `{col_name}` {col_type}")
    
    # 生成ORDER BY子句
    if order_by_columns and len(order_by_columns) > 0:
        order_by_clause = f"ORDER BY ({', '.join(order_by_columns)})"
    else:
        order_by_clause = "ORDER BY tuple()"
    
    # 生成完整的建表语句
    tag_definitions = ',\n'.join(column_definitions)
    create_sql = f"""CREATE TABLE {table_name} (
{tag_definitions}
) ENGINE = {engine}()
{order_by_clause};"""
    
    return create_sql
```

```python
def create_table_from_schema(client, schema_info, table_name, order_by_columns=None):
    """
    仅创建ClickHouse表
    
    Parameters:
    client: ClickHouse客户端
    schema_info: list - schema信息
    table_name: str - 表名
    order_by_columns: list - 排序列
    
    Returns:
    bool - 是否成功
    """
    try:
        # 1. 删除旧表（如果存在）
        # 警告：此操作会删除现有表及其数据，请谨慎使用！
        print(f"正在删除旧表 {table_name}（如果存在）, 确认继续？")
        confirm = input("输入 'y' 继续，其他键取消: ")
        if confirm.lower() != 'y':
            print("操作已取消")
            return False

        drop_sql = f"DROP TABLE IF EXISTS {table_name}"
        client.command(drop_sql)
        print(f"✓ 删除旧表 {table_name}（如果存在）")
        
        # 2. 生成并执行建表语句
        create_sql = generate_create_table_from_schema(schema_info, table_name, order_by_columns)
        print("建表语句:")
        print("-" * 60)
        print(create_sql)
        print("-" * 60)
        
        client.command(create_sql)
        print(f"✓ 创建表 {table_name} 成功")
        
        return True
        
    except Exception as e:
        print(f"❌ 创建表失败: {e}")
        return False

def insert_data_to_table(client, df, table_name):
    """
    仅向已存在的表插入数据
    
    Parameters:
    client: ClickHouse客户端
    df: pandas.DataFrame - 要插入的数据
    table_name: str - 表名
    
    Returns:
    bool - 是否成功
    """
    try:
        print(f"准备插入 {len(df)} 行数据到表 {table_name}...")
        
        # 直接插入数据，不做任何预处理
        client.insert_df(table_name, df)
        print(f"✓ 数据插入成功")
        
        # 验证插入结果
        result = client.query(f"SELECT COUNT(*) as count FROM {table_name}")
        row_count = result.result_rows[0][0]
        print(f"✓ 表中共有 {row_count} 行数据")
        
        return True
        
    except Exception as e:
        print(f"❌ 数据插入失败: {e}")
        import traceback
        traceback.print_exc()
        return False
```

```python
if create_success:
    print("\n" + "=" * 60)
    print("2. 插入数据")
    print("=" * 60)
    
    # 再插入数据
    insert_success = insert_data_to_table(
        client=client,
        df=cleaned_df,
        table_name="gaokao2025"
    )
    
    if insert_success:
        print("\n🎉 表创建和数据插入都完成了！")
    else:
        print("\n❌ 数据插入失败")
else:
    print("\n❌ 表创建失败")
```
